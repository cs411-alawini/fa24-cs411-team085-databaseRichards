```python
from pandas import DataFrame, read_csv
from collections import defaultdict

print(df.columns)

df = read_csv('raw_scores.csv')
df
```

```python
teams = {}

games = defaultdict(list)
for row in df.itertuples():
    date = row[1]
    gameID = row.GAME_ID
    teamID = row.TEAM_ID
    teamName = row.TEAM_ABBREVIATION
    score = row.PTS
    games[gameID].append((date, teamID, teamName, score))
    teams[teamID] = teamName


i = 10
for game in games:
    if i == 0:
        break
    i -= 1
    print(games[game])
    for date, teamID, teamName, score in games[game]:
        print(f'{teamName} {score}')
    print()
```

```python
teamsTable = {"TeamID": [], "TeamName": []}
for teamID, teamName in teams.items():
    teamsTable["TeamID"].append(int(teamID))
    teamsTable["TeamName"].append(teamName)
DataFrame(teamsTable).to_csv('teamsTable.csv', index=False)


betTypesTable = {
  "BetTypeID": [1, 2, 3, 4],
  "BetTypeName": ["Home", "Away", "Over", "Under"],
}

DataFrame(betTypesTable).to_csv('betTypesTable.csv', index=False)
```
