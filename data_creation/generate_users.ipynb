```python
import pandas as pd 
df = pd.read_csv("/Users/nishk/Downloads/334-MB-Test-CSV.csv")
```

```python
df.head()
```

```python
df["First and Last Name"]
df = df[1:]
df.reset_index(drop=True, inplace=True)
```

```python
df["Username"] = df["First and Last Name"].apply(lambda x: ".".join(str(x).lower().split()))
```

```python
df = df[["First and Last Name", "Username"]][:2000]
```

```python
import random
import string
import hashlib
import random
```

```python
passwords = []
for i in range(df.shape[0]):
    passwords.append(''.join(random.choices(string.ascii_uppercase + string.digits, k=6)))
```

```python
df["Password"] = passwords
```

```python
df
```

```python
email_providers = ['@gmail.com', '@outlook.com', '@aol.com', '@hotmail.com', '@yahoo.com']
```

```python
df['Email'] = df['Username']
df['UserID'] = df['Username']
for i in range(df.shape[0]):
    random_number = random.randint(0, 10000)
    df['Email'][i] += str(random_number) + random.choice(email_providers)
    df['UserID'][i] = i
df
```

```python
df = df.drop(columns=['First and Last Name'])
```

```python
df = df[['UserID', 'Username', 'Email', 'Password']]
```
